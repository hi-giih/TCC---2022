# Treinando dados

## Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
SEED = 50

np.random.seed(50)
```

```python
pca_pipeline= Pipeline([('scaler', StandardScaler()), ['PCA', PCA(n_components=2, random_state=SEED)]])
```

### Dados Totais

```python
Dados.shape #linhas por colunas
```

```python
Dados_embbeding_pca= pca_pipeline.fit_transform(Dados) #Mudar tudo pra ponto
```

```python
projectionDT = pd.DataFrame(columns=['x','y'], data=Dados_embbeding_pca)
```

```python
projectionDT
```

# *Kmeans*

```python
import numpy as np #para manipular os vetores
from matplotlib import pyplot as plt #para plotar os gráficos
from sklearn.cluster import KMeans #para usar o KMeans
import  plotly.graph_objects as go 
```

### Dados Totais

```python
kmeans= KMeans(n_clusters= 4, #numero de clusters
init = 'k-means++', n_init = 10, #algoritmo que define a posição dos clusters de maneira mais assertiva
verbose=True, #Explicação 
random_state=SEED)#Aleatoriadade 
```

```python
kmeans.fit(projectionDT)
```

```python
Dados['cluster_pca'] = kmeans.predict(projectionDT)
projectionDT['cluster_pca'] = kmeans.predict(projectionDT)
```
